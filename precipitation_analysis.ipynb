```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import pandas as pd

df_precipitation_file = pd.read_csv('pune_1965_to_2002.csv')
```

# Leitura de datasets

```python
indice = list(df_precipitation_file.Year)
colunas = df_precipitation_file.columns[1:]
valores = df_precipitation_file.values[:,1:]

df_precipitation = pd.DataFrame(valores, index=indice, columns=colunas)
df_precipitation
```

# Agregações

```python
df_precipitation.sum()
df_precipitation.mean(axis='columns')
#overview do dataframe de acordo com o mês
df_precipitation.describe().round(1)
```

# Overview do dataframe de acordo com o ano

```python
df_precipitation.T.describe().round(1)
```

# Precipitações acumuladas segundo o ano 

```python
mpl.style.use('Solarize_Light2')
x = df_precipitation.index
y = df_precipitation.sum(axis='columns')
fig, ax = plt.subplots(figsize=(12,4))
ax.plot(x, y, 'o-', label='Precipitações acumuladas')
ax.set_xlabel('Ano')
ax.set_ylabel('Precipitação acumulada (mm.)')
ax.set_title('Precipitações acumuladas segundo o ano')
ax.legend()
ax.set_xlim(1965, 2002)
#fig.savefig("precipitations_year.pdf")
```

# Precipitações Agosto-Setembro

```python
fig, ax = plt.subplots(figsize=(10,6))
color_mapping = ax.scatter(df_precipitation['Aug'], df_precipitation['Sep'], c=df_precipitation.index)
fig.colorbar(color_mapping)
ax.set_title('Precipitações Agosto-Setembro')
ax.set_xlabel('Precipitações em Agosto (mm.)')
ax.set_ylabel('Precipitações em Setembro (mm.)')
```

# Meses com menor ou maior volume de chuva

```python
fig, ax = plt.subplots(figsize=(8,4))
cumulative_precipitation = df_precipitation.sum()
ax.bar(df_precipitation.columns, cumulative_precipitation)
ax.set_title('Precipitações acumuladas desde 1965 a 2002, segundo o mês')
ax.set_ylabel('Precipitação total (mm.)')
ax.set_xlabel('Mês')
```

# Histograma de Precipitações

```python
fig, ax = plt.subplots(figsize=(8,4))   
ax.hist(df_precipitation.values.flatten(), bins=10)
ax.set_title('Histograma de Precipitações')
ax.set_xlabel('Intervalos de Precipitações (mm.)')
ax.set_ylabel('Frequência absoluta')
```

# Precipitações de Janeiro a Fevereiro

```python
max_january = df_precipitation['Jan'].max()
max_february = df_precipitation['Feb'].max()

fig, ax = plt.subplots(figsize=(12,3))
ax.plot(df_precipitation.index, df_precipitation['Jan'], label='Precipitações de Janeiro')
ax.plot(df_precipitation.index, df_precipitation['Feb'], label='Precipitações de Fevereiro', color='C4')

ax.axhline(max_january,   color='red', linestyle='--', alpha=0.5,  linewidth=3, label='Máxima de janeiro')
ax.axhline(max_february, color='red', linestyle=':', alpha=0.5,   linewidth=3, label='Máxima de fevereiro')

ax.set_xlabel('Ano')
ax.set_ylabel('Precipitação (mm.)')
ax.set_title('Precipitações de janeiro e fevereiro')
ax.set_xlim(df_precipitation.index[0], df_precipitation.index[-1])
ax.legend()
```
